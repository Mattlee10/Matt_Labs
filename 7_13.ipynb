```python
import pandas as pd

# 가상의 사용자 행동 데이터
data = {
    "날짜": ["2025-07-13"],
    "수면 시간 (시간)": [5],
    "음주 여부": ["O"],
    "카페인 섭취": ["있음 (2잔)"],
    "산책 여부": ["X"],
    "운동 여부": ["X"],
    "스트레스 상황": ["업무 스트레스"]
}

df = pd.DataFrame(data)
df
```

```python
def detect_behavior_meaning(row):
    messages = []

    # 스트레스 + 음주
    if row['스트레스 상황'] != "" and row['음주 여부'] == 'O':
        messages.append("스트레스 해소 시도: 음주")

    # 수면 부족 + 카페인
    if row['수면 시간 (시간)'] < 6 and '있음' in row['카페인 섭취']:
        messages.append("스트레스 대응 행동 감지: 수면 부족 + 카페인")

    # 산책 없음 + 운동 없음
    if row['산책 여부'] == 'X' and row['운동 여부'] == 'X':
        messages.append("신체적 회복 행동 부족")

    return messages if messages else ["특이 행동 없음"]
```

```python
df['행동 의미 해석'] = df.apply(detect_behavior_meaning, axis=1)
df[['날짜', '행동 의미 해석']]
```

```python
def generate_sensory_message(meanings):
    ui_feedback = []

    for meaning in meanings:
        if "음주" in meaning:
            ui_feedback.append("🔴 화면이 붉게 물듭니다. 긴장이 느껴집니다.")
        elif "수면 부족 + 카페인" in meaning:
            ui_feedback.append("🟡 붉은 기류가 남아있고, 반응 속도가 살짝 느려집니다.")
        elif "신체적 회복 행동 부족" in meaning:
            ui_feedback.append("⚪️ 무채색의 화면, 흐릿한 인터페이스. 움직임이 멈춰있는 느낌입니다.")
        else:
            ui_feedback.append("🟢 화면이 안정적입니다.")

    return " / ".join(ui_feedback)
```

```python
df['UI 감각 메시지'] = df['행동 의미 해석'].apply(generate_sensory_message)
df[['날짜', '행동 의미 해석', 'UI 감각 메시지']]
```
